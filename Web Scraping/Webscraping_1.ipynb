```python
from nltk.stem import PorterStemmer
# from nltk.tokenize import word_tokenize - having issue with punckt
```

```python
# Used to generate requests from the website we are trying to scrap
import requests
# Library used to scrap the website data
import bs4 as bs
# Used to  is used to automate web browser interaction
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
# Storing the url
url = 'https://forums.t-nation.com'
```

```python
# Creating a new Chrome Session for Selenium
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(executable_path="./chromedriver.exe", options=options)
driver.implicitly_wait(2)
driver.get(url)
```

```python
# Scraping the whole page 
page = driver.execute_script('return document.body.innerHTML')
soup = bs.BeautifulSoup(''.join(page), 'html.parser')
print(soup.text)
```

```python
import pandas as pd
```

```python
# scraping category name
category_col = list()
category = soup.find_all("span", {"class": "category-name"})
# category = soup.find("span", class = "category-name")
#mydivs = soup.findAll("div", {"class": "stylelistrow"})
for cat in category:
    category_col.append(cat.get_text())
    print(cat.get_text())
```

```python
# scraping category description
description_col = list()
cat_desc = soup.find_all("div", {"class": "category-description"})
for desc in cat_desc:
    description_col.append(desc.get_text())
    print(desc.get_text())
```

```python
# scraping forum links
forum_col = list()
forum = soup.find_all("a", {"class": "title"})
for title in forum:
    forum_col.append(desc.get_text())
    print(title.get_text())
```

```python
cat_desc_forum = list((category_col,description_col,forum_col))
#cat_desc_forum = list(map(cat_desc_forum,zip(*l)))
# cat_desc_forum[1].append(category_col)
# cat_desc_forum[2].append(description_col)
# cat_desc_forum[3].append(forum_col)
# print(cat_desc_forum)
df = pd.DataFrame(cat_desc_forum)
print(df)
```

```python
# printing dataframe row 0 all columns
print(df.iloc[0,:])
```

```python
# preprocessing only the forum categories for now
words = []
for i in df.iloc[0,:]:
    group = i.lower().split()
    for j in group:
        words.append(j)
# print(words)
# creating word_set
word_set = set(words)
print(word_set)
```

```python
# Stemming
ps = PorterStemmer()
new_word_set = set()
for w in word_set:
    new_word_set.add(ps.stem(w))
print(new_word_set)
```
